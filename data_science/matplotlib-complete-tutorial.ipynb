## **Pyplot vs Object Oriented Method**

```python
heart= pd.read_csv('/kaggle/input/heart-disease-dataset/heart.csv')
heart
```

```python
over_40 =heart[heart['age']>40] 
over_40
```

```python
# pyplot method

over_40.plot(kind='scatter',
            x= 'age',
            y= 'chol',
            c= 'target');
```

```python
# Object oriented approach

fig,axes =plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=axes);
```

# **Entire Cycle of Object Oriented Plot**

```python
#create figure and graph area

fig,ax= plt.subplots(figsize=(10,6))
#home, room
#plot the data

scatter_plot = ax.scatter(x=over_40['age'],
                         y=over_40['chol'],
                         c=over_40['target'])
# customization
ax.set(title='Suspectability of Heart Diseaser',
      xlabel= 'Age',
      ylabel='Cholestrol')
# set the legend
ax.legend(*scatter_plot.legend_elements(), title = 'target')
scatter_plot;
```

# **Entire life cycle of Object Oriented Plot Advanced**

```python
# Create plot area
fig, (plot1, plot2) = plt.subplots(nrows=2,
                                   ncols=1, 
                                   figsize=(10, 10),
                                  sharex = True)

# Add data to plot
scatter_plot = plot1.scatter(x=over_40['age'], 
                             y=over_40['chol'],
                             c=over_40['target'])

# Customization
plot1.set(title='Heart disease Problem with Cholestrol',
          xlabel='Age',
          ylabel='Cholesterol')

# Define the legend for plot1
legend1 = plot1.legend(*scatter_plot.legend_elements(),
                       title='Survived or Not')

# Now let's plot the second plot
scatter_plot = plot2.scatter(x=over_40['age'],
                             y=over_40['trestbps'],
                             c=over_40['target'])

# Customization for the second plot
plot2.set(title='Heart disease Problem with resting Blood Pressure', 
          xlabel='Age', 
          ylabel='Trestbps')

# Define the legend for plot1
legend2 = plot2.legend(*scatter_plot.legend_elements(), 
                       title='Survived or Not')

# Display the plots
plt.show()
```

# **Final Formatted Graph**

```python
# Create plot area
fig, (plot1, plot2) = plt.subplots(nrows=2,
                                   ncols=1, 
                                   figsize=(10, 10),
                                  sharex = True)

# Add data to plot
scatter_plot = plot1.scatter(x=over_40['age'], 
                             y=over_40['chol'],
                             c=over_40['target'],
                            cmap='winter')

# Customization
plot1.set(title='Heart disease Problem with Cholestrol',
          xlabel='Age',
          ylabel='Cholesterol')

# Define the legend for plot1
legend1 = plot1.legend(*scatter_plot.legend_elements(),
                       title='Survived or Not')

# Add the mean line
plot1.axhline(y=over_40['chol'].mean(), linestyle= '--',color ='Green')

### Now let's plot the second plot
scatter_plot = plot2.scatter(x=over_40['age'],
                             y=over_40['trestbps'],
                             c=over_40['target'])

# Customization for the second plot
plot2.set(title='Heart disease Problem with resting Blood Pressure', 
          xlabel='Age', 
          ylabel='Trestbps')

# Define the legend for plot1
legend2 = plot2.legend(*scatter_plot.legend_elements(), 
                       title='Survived or Not')

# Add the mean line
plot2.axhline(y=over_40['trestbps'].mean(), linestyle= '--',color ='Red')

# Customizing Plot Style
plt.style.use('classic')

# Naming the figure
fig.suptitle('Patient Data',fontsize=32,fontweight='bold')

# Display the plots
plt.show()
```
